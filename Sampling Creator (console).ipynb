```python
import numpy as np
import os
import cv2
import sys

        
im_iterator = 0
old_value_H = 0
old_value_L = 0
old_value_S = 0
old_value_A = 0
        
tmp_im = cv2.imread('C:/Users/SergeyChaika/Desktop/1/1.png')
tmp_im_non_rotated = tmp_im.copy()
        
        
def changeValue_H(value):
        hls_im = cv2.cvtColor(tmp_im, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 0] = hls_im[:, :, 0] + value - old_value_H
        tmp_im = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        hls_im = cv2.cvtColor(tmp_im_non_rotated, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 0] = hls_im[:, :, 0] + value - old_value_H
        tmp_im_non_rotated = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        old_value_H = value
        
        cv2.imwrite('tmp_im.jpg', tmp_im)
        
        
def changeValue_L(value):
        hls_im = cv2.cvtColor(tmp_im, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 1] = hls_im[:, :, 1] + value - old_value_L
        tmp_im = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        hls_im = cv2.cvtColor(tmp_im_non_rotated, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 1] = hls_im[:, :, 1] + value - old_value_L
        tmp_im_non_rotated = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        old_value_L = value
        
        cv2.imwrite('tmp_im.jpg', tmp_im)
        
        
def changeValue_S(value):
        hls_im = cv2.cvtColor(tmp_im, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 2] = hls_im[:, :, 2] + value - old_value_S
        tmp_im = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        hls_im = cv2.cvtColor(tmp_im_non_rotated, cv2.COLOR_RGB2HLS)
        hls_im[:, :, 2] = hls_im[:, :, 2] + value - old_value_S
        tmp_im_non_rotated = cv2.cvtColor(hls_im, cv2.COLOR_HLS2RGB)
        
        old_value_S = value
        
        cv2.imwrite('tmp_im.jpg', tmp_im)
        
        
def changeValue_A(value):
        tmp_im = rotate_bound(tmp_im_non_rotated, value)
        
        cv2.imwrite('tmp_im.jpg', tmp_im)
        
        
def rotate_bound(image, angle):
    # grab the dimensions of the image and then determine the
    # center
    (h, w) = image.shape[:2]
    (cX, cY) = (w // 2, h // 2) 
    # grab the rotation matrix (applying the negative of the
    # angle to rotate clockwise), then grab the sine and cosine
    # (i.e., the rotation components of the matrix)
    M = cv2.getRotationMatrix2D((cX, cY), -angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1]) 
    # compute the new bounding dimensions of the image
    nW = int((h * sin) + (w * cos))
    nH = int((h * cos) + (w * sin)) 
    # adjust the rotation matrix to take into account translation
    M[0, 2] += (nW / 2) - cX
    M[1, 2] += (nH / 2) - cY
    # perform the actual rotation and return the image
    return cv2.warpAffine(image, M, (nW, nH))


for i in range(1, 11)
    for j in range(1, )
```
